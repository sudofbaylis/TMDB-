### Correlations üìàüìâ
Nothing too intresting found yet but could there be anything in the tends of the dataframe?

````python
#checking to see if there are any cool trends in the data 
plt.figure(figsize = (10,10)) # size
sns.heatmap(df.corr(numeric_only=True), # showing correlation
           annot = True, #adding labels
           fmt = '.1%', # making the labels show percentage rather than just a decimal 
           mask = np.triu(df.corr(numeric_only=True)), #only shows half the correlations as they are just repeated 
            cmap='RdPu') # colour of the heatmap
plt.title('Correlations in the Dataframe', fontsize=16)
plt.show()
````

Seems obvious that vote count correlates strongly to popularity and revenue. But revenue and budget having such a strong correlation is intresting. Often you hear big budget movies failing so lets see if we can find any outliers to this trend. 

````python
# Scatter plot of the budget of films vs their revenue also showing a trend line.
plt.figure(figsize=(10,6)) # size of graph 
sns.scatterplot(x='budget', y='revenue', data=df, color='purple')
sns.regplot(x='budget', y='revenue', data=df, scatter=False, color='pink', line_kws={"label": "Trend Line"}) # adding trend line
plt.title('Budget vs Revenue with Trend Line', fontsize=16) 
plt.xlabel('Budget', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.show()
````

***Outlier is Avater (2009)*** at the top -- not exactly shocking as everyone knows what a hit it was.  

***Pirates of the Caribbean: On Stranger Tides*** furthest right -- still following the trend so not hugely shocking.
